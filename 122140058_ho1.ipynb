```python
import numpy as np
import matplotlib.pyplot as plt

# Buat array waktu t dari 0 hingga 2 dengan langkah 0.0001
t = np.arange(0, 2, 0.0001)

# Definisikan sinyal-sinyal
y1 = 2 * np.sin(2 * np.pi * 3 * t)
y2 = 1 * np.cos(2 * np.pi * 4 * t + np.pi / 4)
y3 = -1 * np.sin(2 * np.pi * 5 * t + np.pi / 2)
y4 = 0.5 * np.cos(2 * np.pi * 6 * t + np.pi)

# Plot semua sinyal
plt.figure(figsize=(10, 6))

plt.plot(t, y1, label="y1 = 2 * sin(2π * 3 * t)")
plt.plot(t, y2, label="y2 = cos(2π * 4 * t + π/4)")
plt.plot(t, y3, label="y3 = -sin(2π * 5 * t + π/2)")
plt.plot(t, y4, label="y4 = 0.5 * cos(2π * 6 * t + π)")

# Atur tampilan grafik
plt.xlabel("Waktu (t)")
plt.ylabel("Amplitudo")
plt.title("Grafik Gelombang Sinus dan Cosinus")
plt.legend()
plt.grid()

# Tampilkan grafik
plt.show()
```

Fase akan memengaruhi posisi gelombang berdasarkan nilai pergeseran fase tersebut. Apabila pergeseran fase bernilai positif maka gelombang akan bergeser ke kiri. Sedangkan jika fase bernilai negatif maka gelombang akan bergeser ke kanan.

Perbedaan amlitudo akan memengaruhi tampilan dari gelombang berdasarkan besar kecilnya nilai amplitudo. Apabila nilai amplitudo semakin besar, maka puncak dan lembah dari gelombang semakin tinggi. Sedangkan jika nilai amplitudonya kecil, maka gelombang menjadi lebih pendek.

D. Bandingkan sinyal-sinyal dengan pergeseran fase yang berbeda dan diskusikan bagaimana pergeseran fase mempengaruhi tampilan gelombang.
Pergeseran fase tidak akan merubah bentuk dari gelombang, hanya saja akan memengaruhi posisi dari puncak dan lembah. Artinya pergeseran fase akan menentukan di mana gelombang memulai siklusnya.

```python
import numpy as np
import matplotlib.pyplot as plt

# Buat array waktu t dari 0 hingga 2 dengan langkah 0.0001
t = np.arange(0, 2, 0.0001)

# Definisikan sinyal awal
y1 = 2 * np.sin(2 * np.pi * 3 * t)
y2 = 1 * np.cos(2 * np.pi * 4 * t + np.pi / 4)

# Buat sinyal kombinasi
y5 = y1 + y2

# Plot semua sinyal
plt.figure(figsize=(10, 6))

# Plot y1
plt.subplot(3, 1, 1)
plt.plot(t, y1, label="y1 = 2 * sin(2π * 3 * t)", color='b')
plt.legend()
plt.grid()

# Plot y2
plt.subplot(3, 1, 2)
plt.plot(t, y2, label="y2 = cos(2π * 4 * t + π/4)", color='r')
plt.legend()
plt.grid()

# Plot y5
plt.subplot(3, 1, 3)
plt.plot(t, y5, label="y5 = y1 + y2", color='g')
plt.legend()
plt.grid()

# Tampilkan grafik
plt.tight_layout()
plt.show()
```

Kombinasi antara gelombang sinus/cosinus menghasilkan gelombang yang bergantung pada tiga faktor utama, yaitu amplitudo, frekuensi, dan pergeseran fase. Gelombang dengan amplitudo yang lebih besar akan lebih mendominasi sehingga cenderung menyerupai gelombang dengan amplitudo yang lebih tinggi. Apabila  frekuensinya sama, maka hasilnya akan tetap sinus dengan perubahan amplitudo dan fase, sedangkan jika frekuensinya berbeda, maka akan muncul pola interfrensi dan bisa terlihat menyerupai gelombang campuran.
